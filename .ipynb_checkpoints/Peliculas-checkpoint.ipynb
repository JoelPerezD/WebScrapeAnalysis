Web Scrapping con SELENIUM 

```python
%pip install webdriver-manager

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import pandas as pd

# Configurar Selenium (modo sin ventana)
options = webdriver.ChromeOptions()
options.add_argument('--headless')
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# URL de Wikipedia
url = "https://es.m.wikipedia.org/wiki/Anexo:Pel%C3%ADculas_con_las_mayores_recaudaciones"
driver.get(url)

# Encontrar la tabla de películas
tabla = driver.find_element(By.CLASS_NAME, "wikitable")
filas = tabla.find_elements(By.TAG_NAME, "tr")

datos = []
for fila in filas[1:]:
    celdas = fila.find_elements(By.TAG_NAME, "td")
    if celdas:
        texto_celdas = [celda.text.strip() for celda in celdas]
        texto_celdas += [None] * (7 - len(texto_celdas))  # rellena hasta 7 columnas
        datos.append(texto_celdas[:7])  # 7 columnas que trae la tabla

columnas = [
    "Película", "Recaudación mundial",
    "Taquilla (EE.UU.)", "Taquilla (fuera de EE.UU.)",
    "Presupuesto", "Distribuidora(s)", "Año de estreno"
]

df = pd.DataFrame(datos, columns=columnas)
df.insert(0, "N.º", range(1, len(df) + 1))  # agrega índice como columna "N.º" al principio

print(df.head())
df.to_csv("peliculas_recaudacion_output.csv", index=False)
```

```python
# eliminar fila N.º	 ya que no aporta nada
df = df.drop(columns=["N.º"])
```

```python
df.head()
```

```python
# convertimos a columnas numericas  Recaudación mundial ,  Taquilla (EE.UU.) , Presupuesto  

# Limpiar y convertir "Recaudación mundial"
df["Recaudación mundial"] = (
	df["Recaudación mundial"]
	.astype(str)
	.str.replace(r"\s*\(.*\)", "", regex=True)
	.str.replace(" ", "")
	.astype(float)
)

# Limpiar y convertir "Taquilla (EE.UU.)"
df["Taquilla (EE.UU.)"] = (
	df["Taquilla (EE.UU.)"]
	.astype(str)
	.str.replace(r"\s*\(.*\)", "", regex=True)
	.str.replace(" ", "")
	.str.split("-", n=1).str[0]  # Tomar el primer valor si hay un rango
	.astype(float)
)

# Limpiar y convertir "Presupuesto"
df["Presupuesto"] = (
	df["Presupuesto"]
	.astype(str)
	.str.replace(r"\s*\(.*\)", "", regex=True)
	.str.replace(" ", "")
	.str.split("-", n=1).str[0]  # Tomar el primer valor si hay un rango
	.astype(float)
)

# Limpiar y convertir "Taquilla (fuera de EE.UU.)"
df["Taquilla (fuera de EE.UU.)"] = (
	df["Taquilla (fuera de EE.UU.)"]
	.astype(str)
	.str.replace(r"\s*\(.*\)", "", regex=True)
	.str.replace(" ", "")
	.astype(float)
)
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
#pelicuulas de disnet mas recaudadores 
disney_movies = df[df["Distribuidora(s)"].str.contains("Disney", na=False)].sort_values(
    "Recaudación mundial", ascending=False
)
disney_movies
```

```python
# graficamos las dias primmeras 
plt.figure(figsize=(12, 6))
plt.barh(disney_movies["Película"].head(10), disney_movies["Recaudación mundial"].head(10), color='blue')
plt.xlabel("Recaudación mundial (en miles de millones)")
plt.title("Top 10 Películas de Disney por Recaudación Mundial")
plt.gca().invert_yaxis()  # Invertir el eje y para mostrar la película con mayor recaudación en la parte superior
plt.show()
```

```python
#distribuidoras mas recaudadoras
distribuidoras = df["Distribuidora(s)"].str.split(",").explode().str.strip()
distribuidoras_count = distribuidoras.value_counts().head(10)
plt.figure(figsize=(12, 6))
plt.barh(distribuidoras_count.index, distribuidoras_count.values, color='green')
plt.xlabel("Número de películas")
plt.title("Top 10 Distribuidoras por Número de Películas")
plt.gca().invert_yaxis()  # Invertir el eje y para mostrar la distribuidora con más películas en la parte superior
plt.show()
```
